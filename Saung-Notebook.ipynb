## ARRESTS IN THE NEIGHBORHOODS OF PITTSBURGH. 
My part of focus is on analysing the data of arrests in https://data.wprdc.org/dataset/arrest-data/resource/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f .

```python
import matplotlib.pyplot as plt
import csv
from collections import Counter

x = []
y = []

with open('Arrests data.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots)

    for row in plots:
        neighborhood = row[10].strip()
        x.append(neighborhood)

# Count occurrences
neighborhood_counts = Counter(x)

# Extract data for plotting
x = list(neighborhood_counts.keys())
y = list(neighborhood_counts.values())

# Figure size
plt.figure(figsize=(20, 10))

# Plot the data
plt.bar(x, y, color='pink', width=0.8, label="Number of Arrests")
plt.xlabel('Neighborhoods', fontsize=15)
plt.ylabel('Arrests', fontsize=15)
plt.title('Arrest data of Pittsburgh Neighbourhoods',fontsize=20)

# Rotate x-axis labels and adjust spacing
plt.xticks(rotation=90, ha='right', rotation_mode='anchor', fontsize=10)
plt.tight_layout()
plt.legend()
plt.show()

# Print out the neighborhood with the most and least arrests
most_arrest_neighborhood = max(neighborhood_counts, key=neighborhood_counts.get)
least_arrest_neighborhood = min(neighborhood_counts, key=neighborhood_counts.get)

print("Neighborhood with the most arrests:", most_arrest_neighborhood, "with", neighborhood_counts[most_arrest_neighborhood], "arrests")
print("Neighborhood with the least arrests:", least_arrest_neighborhood, "with", neighborhood_counts[least_arrest_neighborhood], "arrests\n")

# Sort neighborhoods based on arrests 
sorted_neighborhoods = sorted(neighborhood_counts.items(), key=lambda pair: pair[1])
print("7 SAFEST NEIGHBOURHOODS IN PITTSBURGH\n")
for neighborhood, arrests in sorted_neighborhoods[:7]:
    print(f"{neighborhood}: {arrests} arrests")
```

**According to the graph above, Mt. Oliver Neighborhood is the safest in Pittsburgh. But this is only based on the number of people arrested in each neighborhood. To make this more accurate according to the data given in the website, another factor to consider is the number of offences each person commited. So, the next graph is based on this.**


```python
import matplotlib.pyplot as plt
import csv

offenses_per_person = []

with open('Arrests data.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots) 
    for row in plots:
        offenses = row[8].split('/') if len(row) > 8 else [] 
        offenses_per_person.append(len(offenses))

# Counting each number of offenses
offenses_counts = {}
for count in offenses_per_person:
    offenses_counts[count] = offenses_counts.get(count, 0) + 1

# Sorting
sorted_counts = sorted(offenses_counts.items())

# Plotting the data
plt.bar([str(count) for count, freq in sorted_counts], [freq for count, freq in sorted_counts], color='skyblue')
plt.xlabel('Number of Offenses per Person')
plt.ylabel('Number of People')
plt.title('Distribution of Offenses per Person',fontsize=20)

plt.show()

```

**This graph shows the number of offenses commited by each person ranging from 1 to 13. So here comes my next point, it's important to consider that even in areas with low arrest rates, individuals committing multiple offenses contribute significantly to the true crime rate. The final graph combines information from the first two graphs.**

```python
import matplotlib.pyplot as plt
import csv
from collections import Counter

# Read offenses from the CSV file
offenses_per_person = Counter()

with open('Arrests data.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots) 
    for row in plots:
        neighborhood = row[10].strip()
        offenses = row[8].split('/') 
        offenses_per_person[neighborhood] += len(offenses)

# Figure size
plt.figure(figsize=(30, 10))

# Plotting the data and labels
plt.bar(offenses_per_person.keys(), offenses_per_person.values(), color='midnightblue', width=0.8, label="Total Offenses")
plt.xlabel('Neighborhoods',fontsize=20)
plt.ylabel('Total Number of Offenses',fontsize=20)
plt.title('Total Number of Offenses in Pittsburgh Neighborhoods',fontsize=20)

# Adjusting graph
plt.xticks(rotation=90, ha='right', rotation_mode='anchor', fontsize=15)
plt.tight_layout()
plt.legend()
plt.show()

# Find the neighborhood with the most and least offenses
most_offenses_neighborhood = max(offenses_per_person, key=offenses_per_person.get)
least_offenses_neighborhood = min(offenses_per_person, key=offenses_per_person.get)

# Print the neighborhoods with the most and least offenses
print("Neighborhood with the most offenses:", most_offenses_neighborhood, "with", offenses_per_person[most_offenses_neighborhood], "offenses")
print("Neighborhood with the least offenses:", least_offenses_neighborhood, "with", offenses_per_person[least_offenses_neighborhood], "offenses\n")

# Sort neighborhoods based on offenses and print the least 7
sorted_neighborhoods = sorted(offenses_per_person.items(), key=lambda pair: pair[1])
print("7 SAFEST NEIGHBOURHOODS IN PITTSBURGH\n")
for neighborhood, offenses in sorted_neighborhoods[:7]:
    print(f"{neighborhood}: {offenses} offenses")
```

**As I combined the two graphs, I have now got the true data of the total number of offenses commited by each neighborhoods. By doing so, I can tell that Mt. Oliver Neighborhood is truely the safest nieghborhood in terms or offenses/crime/arrest rates.**
