Our project examines neighborhood safety and emergency response efficacy through data analytics, with my part focusing on fire incident data analysis from the [ Pittsburgh Bureau of Fire](https://data.wprdc.org/dataset/fire-incidents-in-city-of-pittsburgh/resource/8d76ac6b-5ae8-4428-82a4-043130d17b02?inner_span=True).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# Load the dataset
url = "https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02"
data = pd.read_csv(url)

# Count the occurrences of each neighborhood
neighborhoods = data['neighborhood'].value_counts()

# Count the occurrences of given incidents
vehicleFires = data[data['type_description'] == 'Passenger vehicle fire']['neighborhood'].value_counts()
buildingFires = data[data['type_description'] == 'Building fire']['neighborhood'].value_counts()

# Set the figure size
plt.figure(figsize=(16, 8))

# Define a custom colormap resembling fire colors for Total Fires
colorGradient = plt.cm.autumn(np.linspace(0, 1, len(neighborhoods)))
neighborhoods.plot(kind='bar', color=colorGradient, edgecolor='black', linewidth=.5, label='Total Fires')

# Draw a wick-like shape at the top of each bar with triangles and a gradient
for i, count in enumerate(neighborhoods):
    plt.plot([i - 0.2, i, i + 0.2], [count, count + 0.6, count], color=colorGradient[i], linewidth=2)

# Reindex vehicleFires to match the neighborhood order
vehicleFires = vehicleFires.reindex(neighborhoods.index, fill_value=0)

# Plot Vehicle Fires
vehicleFires.plot(kind='bar', color='purple', edgecolor='black', linewidth=.5, label='Vehicle Fires')

# Convert buildingFires to a NumPy array and add it to vehicleFires
bottomValues = np.array(vehicleFires)
buildingValues = np.array(buildingFires.reindex(neighborhoods.index, fill_value=0))

# Plot Building Fires on top of Vehicle Fires with the same width
plt.bar(neighborhoods.index, buildingValues, bottom=bottomValues, color='pink', edgecolor='black', width=.5, linewidth=.5, label='Building Fires')

plt.title('Fire Incidents in Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Count')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

The data presented above provides an overview of all fire incidents in a specific neighborhood. Although the raw numbers indicate that Mount Oliver Borough is the safest neighborhood with the lowest number of fire incidents, it's important to distinguish between incidents that pose an actual safety threat and those that are relatively harmless, such as minor barbecues generating excess smoke. To address this issue, I created a proportional bar graph that explicitly highlights fire incidents related to building or vehicle fires. With most calls in a neighborhood being benign, a second graph can clearly display the frequency of fires that pose more urgent safety concerns within the community.

```python
# Combine counts of vehicle and building fires
combined_fires = vehicleFires.add(buildingFires, fill_value=0)

# Sort the combined_fires DataFrame from greatest to smallest
combined_fires = combined_fires.sort_values(ascending=False)

# Set the figure size
plt.figure(figsize=(16, 8))

# Plot Combined Fires
combined_fires.plot(kind='bar', color='purple', edgecolor='black', linewidth=.5, label='Combined Fires')

plt.title('Combined Vehicle and Building Fires')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.legend()

plt.tight_layout()

# Show the plot
plt.show()
```

By excluding isolated, contained, or inconsequential incidents, the refined representation provides a more precise and reliable depiction of genuine safety concerns related to vehicle and building fires. According to the new graph of incidents, New Homestead is the safest neighborhood to reside in, with a significantly lower number of vehicle and building fire incidents than other neighborhoods. While only ranking twelfth in the previous graph when considering all fire incidents.
